```python
# importing packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.array([1.0,3.0])          # size
y_train=np.array([300,700])          # price
print(f"x_train : {x_train}")
print(f"y_train : {y_train}")
```

```python
m=x_train.shape[0]
print(f"No . of training examples : {m}")
```

```python
# training examples
for i in range(2):
    x_i=x_train[i]
    y_i=y_train[i]
    print(f"x_{i}={x_i} , y_{i}={y_i}")
```

```python
# weight and bias
w=200
b=100
print(f"Weight = {w} , Bias = {b}")
```

```python
def compute_output(x,w,b):
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=(w*x[i])+b
    return f_wb
```

```python
f_wb=compute_output(x_train,w,b)
print(f_wb)
```

```python
# plotting 
plt.scatter(x_train,y_train,c="b",marker="*",label="actual values")
plt.plot(x_train,f_wb,c="g",label="Predicted")
plt.title("Housing Prices")
plt.xlabel("size in 1000's sq.ft")
plt.ylabel("price in 1000's of dollars")
plt.legend()
```

```python
# inference
x_test=int(input("Enter the size of house (1000's sq.ft) = "))
f_wb=(w*x_test)+b
print(f"Cost = {f_wb} thousand dollars")
```
